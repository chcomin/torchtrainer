Example of creating an image grid with titles

```python
import matplotlib.pyplot as plt
import skimage
import torch

from torchtrainer.util.img_util import create_grid

# Load some example image
img = skimage.data.cat()
# Size in pixels of each tile in the grid
container_shape = (200,300)

# Create list of 23 images and titles
tensors = []
texts = []
for i in range(23):
    # Put channels on dimension 0 and convert to tensor
    tensors.append(torch.from_numpy(img.transpose(2, 0, 1)))
    texts.append(f"Image {i+1}")

# Create grid with 5 images per row
grid = create_grid(tensors, 5, container_shape, texts)

fig, ax = plt.subplots(figsize=(12,12))
# Show image
ax.imshow(grid)
```
