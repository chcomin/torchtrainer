This notebook shows how to plot all model activations, parameters, gradients and buffers, which is useful for checking if the values are well behaved.

Notice that the activations plots below is not a line plot. This is because the same module might be used multiple times inside a model, which would lead to an ugly plot.

```python
import torch
from torchvision.models import ResNet50_Weights, resnet50

from torchtrainer.util.model_debug import debug_model_state

# Dummy data. In a real application this should be a batch of the dataset.
batch = torch.rand(8, 3, 224, 224)
labels = torch.randint(0, 1000, (8,))
loss_func = torch.nn.CrossEntropyLoss()

weights = ResNet50_Weights.DEFAULT
model = resnet50(weights=weights)
stats_act, stats_params, stats_grads, stats_buffers = debug_model_state(
                                                        model, batch, labels, loss_func)
```
